```python
### filling misssing values with the help of mean of column
df['modelyear_fillNA']=df['Model Year'].fillna(df['Model Year'].mean())
df
```

```python
df=df.rename(columns={'MAKE': 'brand'})
```

```python
df.head()
```

```python
df['newyear']=df['Model Year'].astype(float)
df
```

```python
df['New Electric Range']=df['Electric Range'].apply(lambda x:x*2)
df.head()
```

```python
 #data aggregatting and grouping
df.head()
```

```python
grouped_mean=df.groupby('County')['Model Year'].mean()
```

```python
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['County','City'])['Model Year'].sum()
print(grouped_sum)
```

```python
## aggregate multiple functions
groudped_agg=df.groupby('City') ['Model Year'].agg(['mean', 'sum', 'count'])
groudped_agg
```

```python
### Merging and joining Dataframes
# Create sample DataFrames
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Valuel': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on="Key", how="inner")
```

```python
pd.merge(df1, df2, on="Key", how="outer")
```

```python
pd.merge(df1, df2, on="Key", how="left")
```

```python
pd.merge(df1, df2, on="Key", how="right")
```
